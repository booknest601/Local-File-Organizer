```python
import os

# Create the TTS import file in .txt format

tts_script = """More Than Motivation: 3 Habits That Build Creative Momentum

Ever feel like motivation disappears the moment you actually need it? ...
Yeah, me too. ...
But here’s the thing — creative momentum isn’t about chasing motivation. ...
It’s about building habits that keep you moving, even when that spark feels completely gone. ...

Today, I’ll share three simple habits that help me — and can help you — stay creative, consistent, and confident. ...
Let’s dive in. ...

Habit 1 — Plan for Progress, Not Perfection
Perfection is a trap — and trust me, every creative person falls into it. ...
You spend hours trying to make something flawless... and sometimes, it never even gets finished. ...

The trick? Focus on progress. ...
Each small win counts. Every draft, every idea, every experiment — it all builds momentum. ...

Consistency beats perfection every single time. ...
So instead of waiting for the perfect moment to start, start small — and let progress become your new standard. ...

Habit 2 — Show Up and Speak Out
Confidence doesn’t come first. ...
It’s built every time you show up. ...

So, join the discussion. Ask questions. Share your ideas — even the messy ones. ...
Here’s a secret: people don’t remember the perfect comment; they remember the real one. ...

Every time you speak up, you’re not just sharing — you’re connecting. ...
Your peers are part of your creative ecosystem. The more you engage, the stronger you grow. ...
So don’t hide behind hesitation. Your voice adds value — and sometimes, that’s exactly what someone else needs to hear today. ...

Habit 3 — Turn Feedback Into Growth
Feedback can feel personal — but it’s really fuel for improvement. ...
Think of it like sunlight for your ideas. ...

At first, it can sting — but that’s how growth works. ...
When you treat critique as collaboration, everything changes. ...
You start to see feedback not as judgment... but as partnership. ...

Each note, each suggestion, each comment is an opportunity to learn something new about your work — and about yourself. ...
So instead of running from it, lean in. ...
Use it to refine your craft, strengthen your message, and grow your creative confidence. ...

Closing Reflection
Creative success isn’t about waiting for motivation to strike. ...
It’s about building momentum through small, intentional habits. ...

Plan for progress. ...
Show up and speak out. ...
And always, always turn feedback into growth. ...

Because your story doesn’t start with inspiration — it starts with action. ...
And that next action? It’s yours. ...

Your story starts with you. ...
"""

tts_path = "/mnt/data/TTS_Script_More_Than_Motivation_Eddy_Martinez.txt"

# Ensure the directory exists
os.makedirs(os.path.dirname(tts_path), exist_ok=True)

with open(tts_path, "w") as f:
    f.write(tts_script)

tts_path
```

```python
with open(tts_path, 'r') as f:
    file_content = f.read()
print(file_content)
```

Now, let's import the `gTTS` library and convert the `file_content` to an audio file. The audio will be saved as `TTS_Script_More_Than_Motivation_Eddy_Martinez.mp3` in the `/mnt/data/` directory.

```python
from gtts import gTTS

# The content of the TTS script is already in the 'file_content' variable.

# Language for the TTS
language = 'en'

# Pass the text and language to the gTTS object
myobj = gTTS(text=file_content, lang=language, slow=False)

# Save the converted audio to a file
audio_path = "/mnt/data/TTS_Script_More_Than_Motivation_Eddy_Martinez.mp3"
myobj.save(audio_path)

print(f"Audio file saved to: {audio_path}")
```

You can now play the generated audio file directly in the notebook:

```python
from IPython.display import Audio, display

# Path to the generated audio file
audio_path = "/mnt/data/TTS_Script_More_Than_Motivation_Eddy_Martinez.mp3"

# Display the audio player
display(Audio(audio_path, autoplay=False))
```
